```python
import requests
import pandas as pd
```

```python
from plyer import notification
from datetime import datetime

def alerta(code,tabela):
    
    HTTPcode = {200: "Sucesso", 201:"Created", 203:"Non Authoritative information", 204:"No Content", 
            400:"Bad request", 401: "Unauthorized", 403: "Forbidden", 404: "Not Found", 409: "Conflict",
            500: "Internal Server Error", 501: "Not Implemented", 502: "Bad Gateway", 503: "Service Unavailable", 504: "Gateway Timeout", 599: "Network timeout"
            }
    mensagem = HTTPcode.get(code)
    data = str(datetime.now()) 
    
    notification.notify(
        title= f"Código{code}",
        message = f"{mensagem} ao carregar {tabela}\n {data}",
        timeout=10
    )
```

```python
#Escolher uma API da lista disponibel

url = "https://brasilapi.com.br/"

#Tabela 1

url1 = "https://brasilapi.com.br/api/banks/v1"
tabela1 = requests.get(url1)
code1 =tabela1.status_code

if code1 ==200: #code 200 é o código de sucesso
    data_json1 = tabela1.json() #se for sucesso, converte o conteúdo da resposta da solicitação HTTP (em formato JSON) para um objeto Python
else:
    alerta(code1,"Bank")

bank = pd.DataFrame(data_json1)
bank.head()
```

```python
#Tabela 2
url2 = "https://brasilapi.com.br/api/pix/v1/participants"
tabela2 = requests.get(url2)
code2 =tabela2.status_code

if code2==200: #code 200 é o código de sucesso
    data_json2 = tabela2.json() #se for sucesso, converte o conteúdo da resposta da solicitação HTTP (em formato JSON) para um objeto Python
else:
   alerta(code2,"pix")

pix = pd.DataFrame(data_json2)
pix.head()
```
